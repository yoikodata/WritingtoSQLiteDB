```python
sql_db_connection = sqlite3.connect(r"C:\Users\14087\Downloads\sqlitestudio-3.3.3\SQLiteStudio\Hotel_Bookings_Data.db")
cursor = sql_db_connection.cursor()

cursor.execute("DROP TABLE IF EXISTS Country")
cursor.execute("DROP TABLE IF EXISTS Meal_Classification")
cursor.execute("DROP TABLE IF EXISTS Market_Segment")
cursor.execute("DROP TABLE IF EXISTS Distribution_Channel")
cursor.execute("DROP TABLE IF EXISTS Reserved_Booking_Type")
cursor.execute("DROP TABLE IF EXISTS Assigned_Booking_Type")
cursor.execute("DROP TABLE IF EXISTS Deposit_Booking_Type")
cursor.execute("DROP TABLE IF EXISTS Customer_Classification")
cursor.execute("DROP TABLE IF EXISTS Reservation_Status_Classification")
cursor.execute("DROP TABLE IF EXISTS Agent_Table")
cursor.execute("DROP TABLE IF EXISTS Company_Table")


create_country_table = """CREATE TABLE Country (
  country_ID int [pk],
  country varchar)
"""

create_meal_classification_table = """CREATE TABLE Meal_Classification (
  meal_ID varchar [pk],
  meal varchar)
"""

create_market_segment_table = """CREATE TABLE Market_Segment (
  market_ID varcahr [pk],
  market_segment varcahr)
"""

create_distribution_channel_table = """CREATE TABLE Distribution_Channel (
  distribution_channel_ID varcahr [pk],
  distribution_channel varchar)
"""

create_reserved_booking_table = """CREATE TABLE Reserved_Booking_Type (
  reserved_booking_ID varchar [pk],
  reserved_booking_type varchar)
"""

create_assigned_booking_table = """CREATE TABLE Assigned_Booking_Type (
  assigned_booking_ID varchar [pk],
  assigned_booking_ID_booking_type varchar)
"""

create_deposit_booking_table = """CREATE TABLE Deposit_Booking_Type (
  deposit_ID varchar [pk],
  deposit_booking_type varchar)
"""

create_customer_classification_table = """CREATE TABLE Customer_Classification (
  customer_ID varchar [pk],
  customer_type varchar)
"""

create_reservation_status_classification_table = """CREATE TABLE Reservation_Status_Classification (
  reservation_status_ID varchar [pk],
  reservation_status varchar)
"""

create_agent_table = """CREATE TABLE Agent_Table (
  agent_ID float [pk],
  agent varchar)
"""

create_company_table = """CREATE TABLE Company_Table (
  company_ID float [pk],
  company varchar)
"""

cursor.execute(create_country_table)
cursor.execute(create_meal_classification_table)
cursor.execute(create_market_segment_table)
cursor.execute(create_distribution_channel_table)
cursor.execute(create_reserved_booking_table)
cursor.execute(create_assigned_booking_table)
cursor.execute(create_deposit_booking_table)
cursor.execute(create_customer_classification_table)
cursor.execute(create_reservation_status_classification_table)
cursor.execute(create_agent_table)
cursor.execute(create_company_table)

sql_db_connection.commit()
sql_db_connection.close()
```
